# Comparison of methods in edgeR script
The difference is in the statistical test 

old: LRT
<br>
new: QLF

## First let's compare the top 100 genes ranked by fold change

```r
library(eulerr)
library(dplyr)
library(ggplot2)
set.seed(1)
```

```r
old <- read.delim("data_venn/Matrix.genes.count.dm6.edgeR_old.top100_by_FC.genelist.tsv", header = T, sep = "\t")
new <- read.delim("data_venn/Matrix.genes.count.dm6.edgeR.top100_by_FC.genelist.tsv", header = T, sep = "\t")
```

```r
common <- old[old$x %in% new$x,]
mat <- c(
    "old" = length(old$x) - length(common),
    "new" = length(new$x) - length(common),
    "old&new" = length(common)
)
fit <- euler(mat, shape = "ellipse")
```

```r
plot(fit, 
     quantities = TRUE,
     fills = c("#ffff33", "#377eb8"),
     main = 'top 100 by FC'
) 
```
