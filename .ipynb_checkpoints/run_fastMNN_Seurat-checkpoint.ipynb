```r
suppressPackageStartupMessages({
    library(Seurat)
    library(SeuratData)
    library(SeuratWrappers)
})
#BiocManager::install("biovizBase")  
set.seed(1234)

# Suppress wanrnings
options(warn=-1)
```

```r
# This notebook explain how to use the batch effect normalization using fastMNN 

# Suppose you have two different datasets and want to use them as one.
pbmc.data <- Read10X(data.dir = "data/filtered_gene_bc_matrices/hg19/")
pbmc3k <- CreateSeuratObject(counts = pbmc.data, project = "pbmc3k", min.cells = 3, min.features = 200)
pbmc3k

pbmc10k.data <- Read10X_h5("data/10k_PBMC_3p_nextgem_Chromium_X_filtered_feature_bc_matrix.h5")
pbmc10k <- CreateSeuratObject(counts = pbmc10k.data, project = "pbmc10k", min.cells = 3, min.features = 200)
pbmc10k
```

```r
pbmc <- merge(x = pbmc3k, y = pbmc10k, add.cell.ids = c("pbmc3k", "pbmc10k"), project = "pbmc13K")
```

```r
pbmc <- NormalizeData(pbmc)
pbmc <- FindVariableFeatures(pbmc)
pbmc <- RunFastMNN(object.list = SplitObject(pbmc, split.by = "orig.ident"))
pbmc <- RunUMAP(pbmc, reduction = "mnn", dims = 1:30)
pbmc <- FindNeighbors(pbmc, reduction = "mnn", dims = 1:30)
pbmc <- FindClusters(pbmc)
```

```r
options(repr.plot.width=10, repr.plot.length=5)
DimPlot(pbmc, group.by = c("orig.ident", "ident"), ncol = 2)
```

```r
options(repr.plot.width=10, repr.plot.length=5)
DimPlot(pbmc, group.by = c("orig.ident", "ident"), split.by = 'orig.ident', ncol = 2)
```
