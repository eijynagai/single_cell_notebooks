# Pipeline to generate Loom file for RNA Velocity analysis

Seurat is a very popular tool used for Single-cell Transcriptomics. The usual Seurat object or matrices used as input for Seurat generated by CellRanger do not contain the necessary information for RNA velocity analysis.

[RNA velocity](https://www.nature.com/articles/s41586-018-0414-6) is estimated using the amounts of spliced and unspliced mRNAs. Common single cell feature counts, such as Cell Ranger, do not use them but it is possible to calculate them using FASTQ files.

This pipeline will help you generate the files in Loom format which can be used further for RNA velocity.
